# Receding Piston

Assume the computational domain is $[0, L]$ with wall boundary conditions on both ends, 
the piston structure is initially at $\frac{L}{2}$ with prescribed motion $x_p(t)$. The flow is of constant density $\rho_0$, zero velocity $u_0 = 0$, and constant pressure $p_0$. 

<img src="../Figs/Piston.png" width="400" />

## Fluid equation
The flow is assumed to be compressible, inviscid, and isentropic (no shock), and is governed by 

\begin{align*}
&\frac{\partial u}{\partial t} + u \frac{\partial u}{\partial x} + \frac{1}{\rho}\frac{\partial P}{\partial x} = 0\\
&\frac{\partial \rho}{\partial t} + u \frac{\partial \rho}{\partial x} + \rho\frac{\partial u}{\partial x} = 0\\
&P = P_0\Big(\frac{\rho}{\rho_0}\Big)^{\gamma}
\end{align*}
here $P$ is the pressure, u is the velocity, and $\gamma$ is the gas constant.

Let denote the sound speed $a(\rho) = K\rho^{(\gamma-1)/2}$ with $K = \sqrt{\gamma P_0/\rho_0^{\gamma}}$, the governing equation can be written in the matrix form
\begin{align*}
&\begin{bmatrix}
u \\
\rho
\end{bmatrix}_t +
\begin{bmatrix}
u & \frac{a^2(\rho)}{\rho}\\
\rho & u
\end{bmatrix}
\begin{bmatrix}
u \\
\rho
\end{bmatrix}_x = 0\\
&\begin{bmatrix}
-u-\frac{2}{\gamma-1}a(\rho) \\
-u+\frac{2}{\gamma-1}a(\rho)
\end{bmatrix}_t +
\begin{bmatrix}
u + a(\rho) & \\
 & u - a(\rho)
\end{bmatrix}
\begin{bmatrix}
-u-\frac{2}{\gamma-1}a(\rho) \\
-u+\frac{2}{\gamma-1}a(\rho)
\end{bmatrix}_x = 0\\
\end{align*}

## Riemann invariants and simple wave assumption

The Riemann invariants are $R_{+} = -u-\frac{2}{\gamma-1}a(\rho)$ and $R_{-} = -u+\frac{2}{\gamma-1}a(\rho)$, they are constant on the characteristics 

$$\Gamma_{\pm} : \frac{dx}{dt} = u \pm a(\rho) $$


The simple wave assumption states, both Riemann invariants are constant on each characteristics. Therefore, both characteristic lines are straight lines.




## Approximated analytical solutions

The initial-boundary-value problem of the fluid is with the following two initial conditions and one boundary condition:
$$\rho(x, 0) = \rho_0,\quad  u(x, 0) = 0(t), \quad u(x_p(t), t) = x_p'(t)$$
And the piston velocity is increasing.
The space time domain $\Omega = \{(x, t): 0 \leq x \leq x_p(t)\}$ of the flow consists of 3 parts

<img src="../Figs/Simple-Wave.png" width="400" />

### static part 

   $$x(t) \leq \frac{L}{2} + (u_0 - a_0)t$$
    
   $$\rho, u, p = \rho_0, u_0, p_0$$
    

### close to piston part
   There is a characteristics $\Gamma_{+}$ from the flow (to the piston)
    
   $$R_{+} = -u-\frac{2}{\gamma-1}a(\rho) =  -\frac{2}{\gamma-1}a_0$$
    
   Therefore, the sound speed at the piston $(x_p(\tau), \tau)$ is $$a(\rho(x_p(\tau), \tau)) = a_0 - \frac{\gamma-1}{2}x'_p(\tau)$$
    
   The only characteristic curve from the piston $(x_p(\tau), \tau)$ to the flow is $\Gamma_{-}$ : 
   \begin{align*} 
   &R_{-} = -u+\frac{2}{\gamma-1}a(\rho) = -x_p'(\tau)+\frac{2}{\gamma-1}a(\rho(x_p(\tau), \tau)) = -2x_p'(\tau) + \frac{2}{\gamma-1}a_0\\
   &x = \big(x_p'(\tau) - a(\rho(x_p(\tau), \tau))\big) (t - \tau) + x_p(\tau) = \big(\frac{\gamma+1}{2}x_p'(\tau) - a_0\big) (t - \tau) + x_p(\tau)
    \end{align*}
    
   To determine the flow states at $(x,t)$, we need to first find $\tau$ by solving 
    $$x = \big(\frac{\gamma+1}{2}x_p'(\tau) - a_0\big) (t - \tau) + x_p(\tau)$$
    
   And then determine the flow states with 
    
   \begin{align*}
    &-u-\frac{2}{\gamma-1}a(\rho) =  -\frac{2}{\gamma-1}a_0\\
    &-u+\frac{2}{\gamma-1}a(\rho) = -2x_p'(\tau) + \frac{2}{\gamma-1}a_0
   \end{align*}
   which leads to 
   $$\rho, \, u = \Big(\frac{a_0 - \frac{\gamma-1}{2}x_p'(\tau)}{K}{K}\Big)^{\frac{2}{\gamma-1}},\,x_p'(\tau)$$ 

### rarefaction fan
   The characteristic curve from the piston $(x_p(0), 0)$ forms the right side of the rarefaction fan
    $$x(t) \leq \big(x_p'(0) - a(\rho(x_p(0), 0))\big) t + x_p(0) = \big(\frac{\gamma+1}{2}x_p'(0) - a_0\big) t + x_p(0)$$
    
   We have 
    $$\frac{L}{2} + (u_0 - a_0)t \leq x(t) \leq \big(\frac{\gamma+1}{2}x_p'(0) - a_0\big) t + x_p(0)$$
    
   And the flow variables are determined by $R_{+}$ from the flow and the slope of $\Gamma_{-}$:
   \begin{align*}
    &-u-\frac{2}{\gamma-1}a(\rho) =  -\frac{2}{\gamma-1}a_0\\
    &\frac{x - \frac{L}{2}}{t} = u - a(\rho)
    \end{align*}
    which are 
   $$\rho,\, u = \Big(\frac{2a_0 - \frac{(\gamma-1)(x - \frac{L}{2})}{t}}{(\gamma+1)K}\Big)^{\frac{2}{\gamma-1}},\, \frac{2}{\gamma+1}(a_0 + \frac{x - \frac{L}{2}}{t}) $$

## Structure forced motion
$$x_p(t) = \frac{L}{2} + \frac{t^4}{8}$$

The numerical solution is obtained by the [embeded boundary method](FSI.jl). And the fluid equation there is the Euler equation. The results match well, which indicates the simple wave assumption and the isentropic assumption are valid for this case. 

```julia
using PyPlot
include("Receding-Piston-Exact.jl")
include("FSI.jl")

function structure(t)
    
    ds = 1.0/8.0 * t^4
    vs = 1.0/2.0 * t^3
    as = 3.0/2.0 * t^2

    return [ds; vs; as]
end

L, N = 2.0, 400
fluid_info = [1.4, 1.225, 0. , 1.0] 
T = 1.0

# Compute exact solution
xx_exact, V_exact = Exact_Solution(T,  L, N, fluid_info, structure)

# Compute numerical solution
ms, cs, ks = 1.0, 0.0, 1.0
structure_info = [ms, cs, ks, 0.0, 0.0, L/2, "FORCED", structure] 
Δt = 0.001
time_info = [Δt, T]
fluid, piston = Solve(L, N, fluid_info, structure_info, time_info)
xx, V = fluid.xx, fluid.V


fig, ax = PyPlot.subplots(ncols=3, nrows=1, figsize=(18, 5))

ax[1].plot(xx_exact, V_exact[1, :], "-", fillstyle = "none", label="Simple wave")
ax[2].plot(xx_exact, V_exact[2, :], "-", fillstyle = "none", label="Simple wave")
ax[3].plot(xx_exact, V_exact[3, :], "-", fillstyle = "none", label="Simple wave")
ax[1].plot(xx, V[1, :], "--o", fillstyle = "none", label="Numerical")
ax[2].plot(xx, V[2, :], "--o", fillstyle = "none", label="Numerical")
ax[3].plot(xx, V[3, :], "--o", fillstyle = "none", label="Numerical")

ax[1].set_xlabel("x")
ax[1].set_ylabel("ρ")
ax[1].legend()
ax[2].set_xlabel("x")
ax[2].set_ylabel("v")
ax[2].legend()
ax[3].set_xlabel("x")
ax[3].set_ylabel("p")
ax[3].legend()
fig.tight_layout()
```

## Structure forced motion
$$ x_p(t) =  \frac{L}{2} + u_0(1 - \cos(\pi f t)) \qquad u_0 = 0.1 \qquad f = 2$$

The numerical solution is obtained by the [embeded boundary method](FSI.jl). And the fluid equation there is the Euler equation. The results match well, which indicates the simple wave assumption and the isentropic assumption are valid for this case. 

```julia
function structure(t)
    u0, f = 0.1, 2
    ds =  u0*(1 - cos(pi*f*t))
    vs =  u0*pi*f*sin(pi*f*t)
    as =  u0*pi^2*f^2*cos(pi*f*t)
    
    return [ds; vs; as]
end

L, N = 2.0, 400
mesh_info  = [L, N]
fluid_info = [1.4, 1.225, 0. , 1.0] 
T = 0.5

# Compute exact solution
xx_exact, V_exact = Exact_Solution(T,  mesh_info, fluid_info, structure)

# Compute numerical solution
ms, cs, ks = 1.0, 0.0, 1.0
structure_info = [ms, cs, ks, 0.0, 0.0, L/2, "FORCED", structure] 
Δt = 0.001
time_info = [Δt, T]
fluid, piston = Solve(mesh_info, fluid_info, structure_info, time_info)
xx, V = fluid.xx, fluid.V


fig, ax = PyPlot.subplots(ncols=3, nrows=1, figsize=(18, 5))

ax[1].plot(xx_exact, V_exact[1, :], "-", fillstyle = "none", label="Simple wave")
ax[2].plot(xx_exact, V_exact[2, :], "-", fillstyle = "none", label="Simple wave")
ax[3].plot(xx_exact, V_exact[3, :], "-", fillstyle = "none", label="Simple wave")
ax[1].plot(xx, V[1, :], "--o", fillstyle = "none", label="Numerical")
ax[2].plot(xx, V[2, :], "--o", fillstyle = "none", label="Numerical")
ax[3].plot(xx, V[3, :], "--o", fillstyle = "none", label="Numerical")

ax[1].set_xlabel("x")
ax[1].set_ylabel("ρ")
ax[1].legend()
ax[2].set_xlabel("x")
ax[2].set_ylabel("v")
ax[2].legend()
ax[3].set_xlabel("x")
ax[3].set_ylabel("p")
ax[3].legend()
fig.tight_layout()
```
